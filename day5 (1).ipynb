```python
a = [2, 4, 6, 8]
b = [1, 3, 7, 9]
res = []

i = 0
j = 0

while i < len(a) and j < len(b):
    if a[i] > b[j]:
        res.append(b[j])
        j += 1
    else:
        res.append(a[i])
        i += 1

print(res)
```

```python
# write a program to calulate sum of digit
m=int(input("enter the number: "))
sum=0
n=m
while(n>0):
    rem=n%10
    sum=sum+rem
    n=n//10
print(sum)
```
