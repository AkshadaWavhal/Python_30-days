```python
arr = [0, 1, 0, 3, 12]

index = 0
for i in range(len(arr)):
    if arr[i] != 0:
        arr[index] = arr[i]
        index += 1

for i in range(index, len(arr)):
    arr[i] = 0

print(arr)
```

```python
nums = [1, 2, 2, 3, 1, 4, 2]

freq = {}

for i in nums:
    if i in freq:
        freq[i] += 1
    else:
        freq[i] = 1

print(freq)

```
