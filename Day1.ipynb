```python
# 2 bubble sort
a=[3,5,1,7]
n=len(a)
for i in range(n-1):
    for j in range(n-i-1):
        if(a[j]>a[j+1]):
            a[j],a[j+1]=a[j+1],a[j]
print("sorted array: ",a)
```

```python
# 1. using the for loop
n=int(input("enter the number: "))
fact=1
for i in range(1,n+1):
    fact=fact*i
print(fact)
```

```python
# 2 using recursion
n=int(input("enter the number: "))
fact=1
def fact(n):
    if(n==0 or n==1):
        return True
    return n*fact(n-1)
print(fact(n))
```

```python
#3 finding factorail using while loop
n=5
fact=1
while(n>0):
    fact=fact*n
    n=n-1
print(fact)
```
