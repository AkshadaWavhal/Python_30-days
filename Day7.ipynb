```python
#sum of digit
m=123
sum=0
while(m>0):
    r=m%10
    sum+=r
    m=m//10
print(sum)
```

```python
m = 123
sum = 0

while m > 0:
    r = m % 10      # get last digit
    sum += r        # add digit to sum
    m = m // 10     # remove last digit

print(sum)
```

```python
# count the inter count
arr = list(map(int, input().split()))
freq = {}
for i in arr:
    freq[i] = freq.get(i, 0) + 1
print(freq)
```

```python
# min and max find
a=list(map(int,input().split()))
min=a[0]
max=a[0]
for i in a:
    if(i>max):
        max=i
    else:
        min=i
print("min",min)
print("max",max)
```

```python
a=[32,6,9,11,3]
print("unsorted array: ")
n=len(a)
for i in range(n-1):
    for j in range(n-i-1):
        if(a[j]>a[j+1]):
            a[j],a[j+1]=a[j+1],a[j]
print(a)
```
